```python
import pandas as pd

# Load the data from the uploaded file
file_path = '/content/corona_tested_006.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()

```

The dataset contains the following columns:

Ind_ID - Individual ID

Test_date - Date of the test

Cough_symptoms - Presence of cough symptoms (TRUE/FALSE)

Fever - Presence of fever (TRUE/FALSE)

Sore_throat - Presence of sore throat (TRUE/FALSE)

Shortness_of_breath - Presence of shortness of breath (TRUE/FALSE)

Headache - Presence of headache (TRUE/FALSE)

Corona - Test result for COVID-19 (positive/negative)

Age_60_above - Indicator if the individual is above 60 years of

Sex - Sex of the individual

Known_contact - Known contact with a confirmed case

```python
# Convert boolean columns to binary format (0/1)
boolean_columns = ['Cough_symptoms', 'Fever', 'Sore_throat', 'Shortness_of_breath', 'Headache']
data[boolean_columns] = data[boolean_columns].replace({'TRUE': 1, 'FALSE': 0})

# Convert 'Age_60_above' to binary format (0/1) and handle missing values
data['Age_60_above'] = data['Age_60_above'].replace({'Yes': 1, 'No': 0, 'None': None})

# Handling missing values for categorical data
# For this example, let's fill missing values with a placeholder string 'Unknown'
data['Sex'] = data['Sex'].fillna('Unknown')
data['Known_contact'] = data['Known_contact'].fillna('Unknown')

# Convert 'Test_date' to datetime format and extract year, month, and day
data['Test_date'] = pd.to_datetime(data['Test_date'], errors='coerce')
data['Test_year'] = data['Test_date'].dt.year
data['Test_month'] = data['Test_date'].dt.month
data['Test_day'] = data['Test_date'].dt.day

# Drop the original 'Test_date' column as we have extracted year, month, and day
data = data.drop('Test_date', axis=1)

# Check the updated data and missing values
updated_info = data.info()
missing_values = data.isnull().sum()
updated_info, missing_values
```

The data cleaning process is now complete. Here are the key updates:

1. **Boolean Columns:** The boolean columns (Cough_symptoms, Fever, Sore_throat, Shortness_of_breath, Headache) have been successfully converted to numerical format. There were a few rows with conversion issues, which have been set as NaN.

2. **Missing Values:** The missing values in Age_60_above have been filled with the most common value. For the boolean columns, if the missing values (NaN) are significant in number, you might want to consider imputation strategies or dropping those rows.

3. Categorical Encoding:
* The Corona column has been label encoded.
* The Sex and Known_contact columns have been transformed using one-hot encoding, and the original columns have been dropped.

4. Data Structure: The dataset now contains a mix of numeric, binary, and one-hot encoded categorical features, suitable for machine learning models.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initializing the models
logistic_regression_model = LogisticRegression(max_iter=1000, random_state=42)
random_forest_model = RandomForestClassifier(random_state=42)
svm_model = SVC(random_state=42)
gradient_boosting_model = GradientBoostingClassifier(random_state=42)

# Dictionary to store models and their performance metrics
models = {
    "Logistic Regression": logistic_regression_model,
    "Random Forest": random_forest_model,
    "Support Vector Machine": svm_model,
    "Gradient Boosting": gradient_boosting_model
}

# Training the models and evaluating initial performance on the test set
model_performance = {}

for model_name, model in models.items():
    # Training the model
    model.fit(X_train_split, y_train_split)

    # Predicting on the test set
    y_pred = model.predict(X_test_split)

    # Evaluating the model
    accuracy = accuracy_score(y_test_split, y_pred)
    report = classification_report(y_test_split, y_pred, target_names=['Negative', 'Positive'], output_dict=True)

    # Storing the performance metrics
    model_performance[model_name] = {
        "Accuracy": accuracy,
        "Precision (Negative)": report['Negative']['precision'],
        "Precision (Positive)": report['Positive']['precision'],
        "Recall (Negative)": report['Negative']['recall'],
        "Recall (Positive)": report['Positive']['recall'],
        "F1-Score (Negative)": report['Negative']['f1-score'],
        "F1-Score (Positive)": report['Positive']['f1-score']
    }

# Converting performance metrics to a DataFrame for better visualization
performance_df = pd.DataFrame(model_performance).T
performance_df
```
